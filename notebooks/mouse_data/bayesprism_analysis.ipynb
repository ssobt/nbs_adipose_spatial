```r
## Notebook env: farnaz_spatial (R kernel)
## this notebook analyzes the decomposition of celltypes from bayesprism and compares it to 
## the decompostion from seurat
```

```r
library(stringr)
library(BayesPrism)
library(ggplot2)
library(dplyr)
```

```r
decovoluted_names = list.files('/home/ssobti/projects/farnaz_spatial/output_data/bayesprism/')
decovoluted_names
```

```r
decovoluted_files = paste0('/home/ssobti/projects/farnaz_spatial/output_data/bayesprism/', decovoluted_names)
decovoluted_objs = lapply(decovoluted_files, readRDS)
```

```r
names(decovoluted_objs) = str_sub(decovoluted_names, start = 1, end = -5) 
```

```r
thetas = lapply(decovoluted_objs, get.fraction, which.theta = "final", state.or.type = "type")
```

```r
head(thetas[[1]])
```

```r
max_confidence_value_per_spot <- lapply(thetas, function(x) apply(x, 1, max))
names(max_confidence_value_per_spot) = names(decovoluted_objs)
```

```r
max_val_df = stack(max_confidence_value_per_spot)
colnames(max_val_df) = c('values', 'slice')
head(max_val_df)
```

```r
options(repr.plot.width=10, repr.plot.height=6)
ggplot(max_val_df, aes(x=values, fill=slice, color=slice)) +
  geom_histogram(position="identity", alpha=0.5, bins = 200) + scale_color_grey() + scale_fill_grey() +
  xlab('Max probability of any celltype per spot') + theme_classic()
```

```r
## R2 comparison to results from Seurat integration
seurat_decompositions = readRDS('/home/ssobti/projects/farnaz_spatial/output_data/seurat/seurat_mBAT_ref_decomposition_scores.rds')
```

```r
seurat_decompositions = lapply(seurat_decompositions, t)
```

```r
head(seurat_decompositions[[1]])
```

```r
seurat_decompositions = lapply(seurat_decompositions, as.data.frame)
max_vals_seurat = lapply(seurat_decompositions, select, 'max')
max_vals_seurat_df = data.frame(slice = unlist(mapply(rep, names(seurat_decompositions), sapply(seurat_decompositions, nrow)), use.names = FALSE), max_val = unlist(max_vals_seurat, use.names = FALSE))
```

```r
### Seurat decompostion max probability for any celltype
options(repr.plot.width=10, repr.plot.height=6)
ggplot(max_vals_seurat_df, aes(x=max_val, fill=slice, color=slice)) +
  geom_histogram(position="identity", alpha=0.5, bins = 200) + scale_color_grey() + scale_fill_grey() +
  xlab('Max probability of any celltype per spot') + theme_classic()
```

```r
seurat_decompositions[[1]] = seurat_decompositions[[1]][ , !(colnames(seurat_decompositions[[1]]) %in% 'max')]
seurat_decompositions[[2]] = seurat_decompositions[[2]][ , !(colnames(seurat_decompositions[[2]]) %in% 'max')]
seurat_decompositions[[3]] = seurat_decompositions[[3]][ , !(colnames(seurat_decompositions[[3]]) %in% 'max')]
seurat_decompositions[[4]] = seurat_decompositions[[4]][ , !(colnames(seurat_decompositions[[4]]) %in% 'max')]
```

```r
cell_type_nms_corrected <- gsub("-", "_", colnames(seurat_decompositions[[1]]))
```

```r
cell_type_nms_corrected
```

```r
colnames(seurat_decompositions[[1]]) <- cell_type_nms_corrected
colnames(seurat_decompositions[[2]]) <- cell_type_nms_corrected
colnames(seurat_decompositions[[3]]) <- cell_type_nms_corrected
colnames(seurat_decompositions[[4]]) <- cell_type_nms_corrected

colnames(thetas[[1]]) <- cell_type_nms_corrected
colnames(thetas[[2]]) <- cell_type_nms_corrected
colnames(thetas[[3]]) <- cell_type_nms_corrected
colnames(thetas[[4]]) <- cell_type_nms_corrected
```

```r
bayesprism_preds <- lapply(thetas, function(x) cell_type_nms_corrected[apply(x, 1, which.max)])
```

```r
names(bayesprism_preds)
```

```r
df = data.frame(cell_type = unlist(bayesprism_preds), slice = unlist(mapply(rep, c('A1','B1','C1','D1'), sapply(bayesprism_preds, length)), use.names = FALSE))
```

```r
count_tbl <- df %>% group_by(slice, cell_type) %>% summarize(n = n())
head(count_tbl)
```

```r
percent_tbl <- df %>% group_by(slice, cell_type) %>% summarize(n = n()) %>% mutate(percent = 100*n / sum(n))
head(percent_tbl)
```

```r
## Frequency of cell types by slice (bayesprism)
set.seed(5)
ggplot(data=count_tbl, aes(x=slice, y=n, fill=cell_type)) + ylab('count') +
geom_bar(stat="identity") + scale_fill_manual(values = sample(scales::viridis_pal(begin = 0.4, end = 0.9, option = 'B', direction = -1)(14), replace = FALSE)) + 
theme_classic() + theme(text=element_text(size=18))
```

```r
## Frequency of cell types by slice (bayesprism)
set.seed(5)
ggplot(data=percent_tbl, aes(x=slice, y=percent, fill=cell_type)) + ylab('percent') +
geom_bar(stat="identity") + scale_fill_manual(values = sample(scales::viridis_pal(begin = 0.4, end = 0.9, option = 'B', direction = -1)(14), replace = FALSE)) + 
theme_classic() + theme(text=element_text(size=18))
```

```r
seurat_preds <- lapply(seurat_decompositions, function(x) cell_type_nms_corrected[apply(x, 1, which.max)])
```

```r
percent_same_celltype_prediction = mapply(FUN = function(x,y) return(100*length(which(x == y))/length(x)), bayesprism_preds, seurat_preds, SIMPLIFY = TRUE)
```

```r
similarity_df = data.frame(slice = names(thetas), percent_similarity = percent_same_celltype_prediction)
similarity_df
```

```r
### BayesPrism and Seurat Integration results for celltype decompostion are extremely similar
options(repr.plot.width = 10, repr.plot.height=8)
ggplot(similarity_df, aes(x=slice, y=percent_similarity, fill=slice)) +
geom_bar(stat="identity", color="black", width = 0.6) + geom_hline(yintercept = 100) +
scale_fill_manual(values=c("#999999", "#E69F00", "#56B4E9", 'black')) + ylab('% of cells BayesPrism and Seurat predict as same celltype') +
  theme_classic() + theme(text=element_text(size=18))
```
